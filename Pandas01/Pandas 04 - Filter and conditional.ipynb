# Pandas 04 - Filter and conditional.
Usar condiciones para filtar por colummas.

```python
import pandas as pd
```

Dataset de Pokemon.

```python
df = pd.read_csv('pokemon.csv')
```

```python
df.head(5)
```

Filtar registros por condicion

```python
# Filtrar pokemons cuando el Type 2 == 'Fire'
df.loc[df.loc[:, 'Type 2'] == 'Fire']
```

```python
# crear una variable de filtrado
fil1 = df['Type 2'] == 'Fire'
df.loc[fil1]
```

### Usando el metodo and &.
Para usar el metodo and, se deben pasar cada condicion encerrada en () y separadas por &

```python
op1 = df.loc[:, 'Type 1'] == 'Fire'
op2 = df.loc[:, 'Generation'] == 1
df.loc[(op1) & (op2)]
```

```python
# lo anterior en una sola linea
df.loc[(df.loc[:, 'Type 1'] == 'Fire') & (df.loc[:, 'Generation'] == 1)]
```

```python
# el anteriro y solo seleccionar determinadas columnas
df.loc[(df.loc[:, 'Type 1'] == 'Ghost') & (df.loc[:, 'Generation'] == 1), ['Name', 'HP', 'Generation']]
```

```python
# la manera simple de arriba seria hacer una variable y encerrar las condiciones ()
con = (df.loc[:, 'Type 1'] == 'Ghost') & (df.loc[:, 'Generation'] == 1)
df.loc[con, ['Name', 'HP', 'Generation']]
```

### Hacer uso del operador or |.
este operador retorna los registros que concida con una u otra condicion.

```python
# traerme el tipo Ghost o pokemons con HP en 100
op2 = (df.loc[:, 'Type 1'] == 'Ghost') | (df.loc[:, 'HP'] ==  100)
df.loc[op2]
```

### Operador not ~.
Este operador retorna los registros cuando no esten en las condiciones que se le pasen.

```python
no1 = (df.loc[:, 'Type 1'] == 'Fire') & (df.loc[:, 'HP'] == 80)
df.loc[no1]
```

```python
# usando el operador not para compararlo con el de arriba
# el filtro devuelve todo cuando no este en Type 1 Fire y en HP = 100
df.loc[~no1]
```

```python
# este devuelve los registros cuando no este en una condicion u otra
no2 = (df.loc[:, 'Type 1'] == 'Fire') | (df.loc[:, 'HP'] == 100)
df.loc[~no2]
```

### Usando el metodo isin().
Metodo para devlver los resgistros cuando se le pase una lista como condicion.

```python
# devuelve todos los registros cuando Type 1 se ecuentre en my_in
my_in = ['Fire', 'Water', 'Ghost']
df.loc[df.loc[:, 'Type 1'].isin(my_in), ['Name', 'Type 1','HP']].to_csv('uno.xlsx')
df.loc[df.loc[:, 'Type 1'].isin(my_in), ['Name', 'Type 1','HP']]
```

```python
# usar and mas isin
myin = ['Ghost', 'Rock', 'Fire']
# la condicion isin va al final
fun2 = (df.loc[:, 'HP'] == 85) & (df.loc[:, 'Type 1'].isin(myin))
df.loc[fun2, ['Name', 'Type 1', 'HP']].to_excel('dos.xlsx')
df.loc[fun2, ['Name', 'Type 1', 'HP']]
```

```python
# usar isin con or
ap = (df.loc[:, 'HP'] == 85) | (df.loc[:, 'Type 1'].isin(myin))
df.loc[ap, ['Name',  'Type 2', 'HP']].to_excel('tres.xlsx')
df.loc[ap, ['Name', 'Type 1', 'HP']]
```

```python
# usar isin con not
myin1 = ['Fire','Water', 'Rock', 'Grass', 'Ghost', 'Psychic']
ap2 = (df.loc[:, 'HP'] == 85) | (df.loc[:, 'Type 1'].isin(myin1))
df.loc[~ap2, ['Name', 'Type 1', 'HP']].to_excel('cuatro.xlsx')
df.loc[~ap2, ['Name', 'Type 1', 'HP']]
```

```python
ap2 = (df.loc[:, 'HP'] == 85) & (df.loc[:, 'Type 1'].isin(myin1))
df.loc[~(ap2), ['Name', 'Type 1', 'HP']]
```
