### Acceder a filas y columnas por indices (solo enteros) con iloc[].
iloc recibe 2 enteros pasados como argumentos, el segundo es opcional, si no se pasa el indice de la columan retorna todas.

```python
# retorna una lista que representa la fila pasada como argumento en iloc
df_csv.iloc[1]
```

```python
# se debe utilizar doble [[]] para retornar un DataFrame
df_csv.iloc[[1]]
# retorna la fila y todas las columnas si no se especifica un segundo argumento el cual
# seria el indice de la columna
```

```python
# acceder a fila y a columna por su indice
df_csv.iloc[[1], [3]]
```

```python
# pasar varios indices y columnas en iloc
df_csv.iloc[[1,2,3], [1,3,5]]
```

```python
# usar rebandas (slices) no llevan doble []
df_csv.iloc[1:4, [1]]
```

```python
# usar slices en fila y columna. Retorna un DataFrame
df_csv.iloc[1:5, 1:3]
```

```python
# si no se especifica el indice de la columna retorna todas
df_csv.iloc[1:5]
```

```python
# seleccionar determindas columnas y mostrar todas las filas
df_csv.iloc[:, [1,2,3]]
```

### Seleccion de columnas y filas con loc[].
loc puede acceder tanto a las filas como a las columnas por el nombre del indice, en caso de tener etiquetas de nombre en los indices de filas, de lo contrario se acceder por el numero entero.

```python
df_csv.loc[[1], ['Employment']]
```

```python
# acceder a determinada columna y mostrar todas las filas
df_csv.loc[:, ['Employment']]
```

### Series de pandas
las series en pandas solo son las columnas, mientras que un DataFrame es la tabla completa con las filas y columnas.

```python
# crear una serie a partir de una lista
lista = [1,2,3,4,5,6,7,8]
lps = pd.Series(lista)
```

esto crea un serie la cual es solo una columna, si no se pasa nigun indice, pandas le asigna uno por defetco

```python
lps
```

```python
# la serie es solo accesible solo por el indices
lps[3]
```

asignar etiquetas a los indices de la serie

```python
lps = pd.Series(lista, index=['a', 'b,', 'c', 'd', 'e', 'f', 'g', 'h'])
```

```python
# Ahora las etiquetas tiene nombre de indice
lps
```

```python
# acceder a una fila por su etiqueta
lps['a']
```

### Crear directamente una serie con etiquetas de indice.

```python
calorias = {'coca': 300, 'jugo': 200, 'natural': 75, 'agua': 65}
cal = pd.Series(calorias)
cal
```

### Metodos adicionales para series.
Tratarlo como un diccionario.

```python
# usar el metodo in con el indice
'jugo' in cal
```

```python
# si un indice no esta produce un keyerror
cal['jugo']
```

devuelve un keyerror al no encontrarse en la serie
cal['soda']

```python
# usar elmetodo get y un argumwnto para devolver una segunda opcion
cal.get('natural', '#n/a')
```

```python
cal.get('soda', '#n/a')
```

### Atributo de nombre.
una serie tambien puede tener un nombre asignado.

```python
cal1 = pd.Series(calorias, name='Serie calorias')
```

### Crear un DataFrame a partir de Series.

```python
d = {
    'one': pd.Series([1.0,2.0,3.0], index=['a','b','c']),
    'two': pd.Series([10,20,30,40], index=['a','b','c','d'])
}
dd = pd.DataFrame(d)
```

```python
# si el largo es diferente por omision en una serie se completa con NaN automaticamente
dd
```

```python
# renombrar columnas
dd.rename(columns={'one': 'Uno', 'two': 'Dos'})
```

```python
# crear DataFrame seleccionando determinadas columnas
de = pd.DataFrame(d, columns=['two'])
```
