# Pandas 07 - Sorting data 

```python
import pandas as pd
```

```python
persons = {
    'First': ['Corey', 'Guido', 'Walter', 'Jesse'],
    'Last': ['Doe', 'Van', 'White', 'Pinkman'],
    'Occupation': ['Developer', 'Master', 'Chamestry', 'Businessman']
}
df = pd.DataFrame(persons)
df
```

### Metodo sort_values()
by toma la columna a ordenar, por defecto ordena de forma ascendente, para cancelar se utiliza la funcion ascending=False.

```python
# ordena los elementos por columna de manera ascendente
df.sort_values(by='Last')
```

```python
# ordena de manera descendente
df.sort_values(by='Last', ascending=False)
```

```python
df.sort_values(by='First')
```

```python
# ordenar varias columnas
df.sort_values(by=['First', 'Last'])
```

```python
df.sort_values(by=['First', 'Last'], ascending=False)
```

### Metodo sum().
Devuelve la suma de todo el eje, por ejemplo todos los valores de una columna.

```python
stack['Respondent'].sum()
```

```python
# sumar todas las edades
stack['Age'].sum()
```

### Metodo nlargest().
Se utiliza para obtener los valores mas grandes. Toma 2 argumentosnel primero es un entero que representa el numeronde registros a mostras y el segundo es la columna a usar.

```python
stack.nlargest(20, 'Age')
```
