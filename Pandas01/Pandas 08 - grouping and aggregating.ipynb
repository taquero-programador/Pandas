```python
# agg, muestra la media y la constante
df.groupby(['Country'])['ConvertedComp'].agg(['median', 'mean'])
```

```python
# igual a lo anterior haciendo uso de loc
df.groupby(['Country'])['ConvertedComp'].agg(['median', 'mean']).loc['Canada']
```

```python
# filtrar pais y todo lo que contenga python
df.loc[df.loc[:, 'Country'] == 'India']['LanguageWorkedWith'].str.contains('Python')
```

```python
df.loc[df.loc[:, 'Country'] == 'India']['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# por cada pais suma la coincidencia por Python
df.groupby(['Country'])['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
df['Country'].value_counts()
```

```python
pdlambda = df.groupby(['Country'])['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
pdlambda
```

```python
crespondes = df['Country'].value_counts()
```

```python
use_python = df.groupby(['Country'])['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
# une dos columnas de 2 diferentes consultas
python_df = pd.concat([crespondes, use_python], axis='columns', sort=False)
python_df
```

```python
# renombras las columnas del nuevo df
python_df = python_df.rename(columns={'Country': 'NumsRespod', 'LanguageWorkedWith': 'Pythonusers'})
```

```python
# esto crea una nueva columna Valnew que es el resultado de la siguiente operacion
python_df['Valnew'] = (python_df['Pythonusers'] / python_df['NumsRespod']) * 100
python_df
```

```python
# ordena los valores y lo guarda
python_df.sort_values(by='Valnew', ascending= False, inplace=True)
```
